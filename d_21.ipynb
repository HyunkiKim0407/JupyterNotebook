```python
x1 =dataset_1['x']
y1 =dataset_1['y']
x2 =dataset_2['x']
y2 =dataset_2['y']
x3 =dataset_3['x']
y3 =dataset_3['y']
x4 =dataset_4['x']
y4 =dataset_4['y']



#회귀선 계산
#두 변수간에 선형관계를 모델링 - 기울기, 절편, 상관계수, 가설검정, 표준오차


from scipy.stats import linregress
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x1, y1)
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(x2, y2)
slope3, intercept3, r_value3, p_value3, std_err3 = linregress(x3, y3)
slope4, intercept4, r_value4, p_value4, std_err4 = linregress(x4, y4)
```

```python
fig = plt.figure() # 도화지 생성
axis1 = fig.add_subplot(2,2,1) #2행 2열짜리의 1번째 따로 도화지
axis2 = fig.add_subplot(2,2,2)
axis3 = fig.add_subplot(2,2,3)
axis4 = fig.add_subplot(2,2,4)

axis1.plot(x1, intercept1+slope1*x1, color='red', label = 'Regression Line')
axis2.plot(x2, intercept2+slope2*x2, color='g', label = 'Regression Line')
axis3.plot(x3, intercept3+slope3*x3, color='b', label = 'Regression Line')
axis4.plot(x4, intercept4+slope4*x4, color='violet', label = 'Regression Line')
```

```python
from scipy.stats import norm
data = np.random.randn(1000)

print(np.mean(data))
np.std(data)#이산형 데이터 표(테이블) #연속형 표로 구현하기 어렵다 > 그래프 > 구간의 넓이 > 확률 밀도 함수를 사용한다. 

plt.hist(data, bins=30, density=True, alpha=0.7,color='pink', edgecolor='r')
xmin, xmax = plt.xlim() #그래프가 출력하는 x축의 최저값, 최상값을 받기
x = np.linspace(xmin, xmax, 100)#xmin, xmax 100의 균등한 데이터 출력
p = norm.pdf(x, 0, 1)#x데이터, 0평균, 1표면, x값에서 확률 밀도 값을 반환(kde)
plt.plot(x,p, 'b', linewidth=2)
plt.title('Standard Normal Distribution and Histogram')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
#################################

def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1


tips['sex_num'] = tips['sex'].apply(recode_sex)
import matplotlib.pyplot as plt

#x, y, c 다변량 데이터 그래프 그리기
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
scatter = axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 20,  #데이터프레임의 'size' 열 값을 20배로 확장하여 마커 크기를 설정
    c=tips['sex_num'],  # 'sex_num'에 따라 다른 색상 지정
    alpha=0.5, cmap='Blues')


plt.colorbar(scatter, label='Sex')

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 



```

```python
# =============================================================================
#  그 중 하나가 'tips' 데이터셋 
#  'tips' 데이터셋은 음식점에서의 식사 정보에 관한 데이터를 포함하고 있으며, 
#  주로 데이터 시각화 및 통계 분석 연습에 사용
#  이 데이터셋은 팁(tip)과 관련된 정보를 포함하고 있으므로 
#  주로 팁과 관련된 분석 및 시각화를 수행할 때 사용
# 
# # 'tips' 데이터셋에 포함된 열(컬럼)은 다음과 같습니다:
# 
# # total_bill: 식사 대금(음식과 음료의 총 비용)을 나타내는 숫자 값
# 
# # tip: 팁(봉사료)을 나타내는 숫자 값 
# 이 값은 고객이 추가로 지불한 금액을 나타냅니다.
# 
# # sex: 고객의 성별을 나타내는 범주형 변수 
# 'Male' 또는 'Female' 값 중 하나를 가집니다.
# 
# # smoker: 흡연 여부를 나타내는 범주형 변수 
# 'Yes' 또는 'No' 값 중 하나를 가집니다.
# 
# # day: 식사가 발생한 요일을 나타내는 범주형 변수 
# 'Thur' (목요일), 'Fri' (금요일), 'Sat' (토요일), 'Sun' (일요일) 값 중 하나를 가집니다.
# 
# # time: 식사 시간을 나타내는 범주형 변수 
# 'Lunch' (점심), 'Dinner' (저녁) 값 중 하나를 가집니다.
# 
# # size: 테이블에 앉은 고객의 수를 나타내는 숫자 값
# 
# # 'tips' 데이터셋은 주로 식사 대금(total_bill)과 팁(tip) 간의 관계, 
# 성별(sex) 및 흡연 여부(smoker)에 따른 팁의 차이, 
# 요일(day)과 식사 시간(time)에 따른 팁의 변화 등을 분석하고 시각화하는 데 
# 사용 Seaborn 라이브러리의 예제 데이터셋 중 하나로, 
# 데이터 분석 및 시각화를 연습하고 테스트하는 데 유용한 자료
```
