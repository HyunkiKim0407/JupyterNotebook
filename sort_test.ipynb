```python
arr = [3,4,5,1,2]
def selection_sort(arr):
    for end in range(len(arr)):      
        for i in range(end + 1, len(arr)):
            if arr[end] > arr[i]:
                arr[end], arr[i] = arr[i], arr[end]
              
selection_sort(arr)
print('##########정답 :', arr)
```

```python
arr = [10, 8, 6, 20, 4, 3, 22, 1, 0, 15, 16]
# arr = [10, 8, 6, 20, 4, 3, 22, 1, 0, 15]
def shell_sorting(arr):
    gap = len(arr) // 2
    while gap > 0:
        
        for i in range(len(arr)):
            for z in range(i + gap, len(arr), gap):
                if arr[i] > arr[z]:
                    arr[i],arr[z] = arr[z],arr[i]
        gap //= 2   # 들여쓰기 잘못되서 틀렸음
    
shell_sorting(arr)
print(arr)
```

```python
arr = [10, 8, 6, 20, 4, 3, 22, 1, 0, 15, 16]
def shell_sorting(arr):
    gap = len(arr) // 2 # 간격을 배열의 길이의 절반으로 초기화
    while gap > 0: # 간격이 0보다 큰 동안 반복
        for i in range(len(arr)): # 배열의 모든 요소에 대해 반복
            for z in range(i + gap, len(arr), gap): # i + gap부터 배열의 끝까지 간격만큼 건너뛰며 반복
                if arr[i] > arr[z]: # 만약 arr[i]가 arr[z]보다 크다면
                    arr[i],arr[z] = arr[z],arr[i] # arr[i]와 arr[z]를 교환
        gap //= 2 # 간격을 절반으로 줄임
                    
shell_sorting(arr) # 쉘 정렬 함수 호출
print(arr) # 정렬된 배열을 출력
```
