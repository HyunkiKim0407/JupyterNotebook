```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3, 0.3)
y = np.arange(0,30, 5)
y1 = x ** 2
y2 = x ** 3

plt.plot(x,x, 'k*', linewidth = 15)
plt.plot(x,y1, 'b^-.')
plt.plot(x,y2, 'ro-', markersize = 15)

plt.grid(axis = 'y', color = 'royalblue', linestyle = '--', alpha = 0.5)

labels = ['1d','2d','3d','4d','5d','6d','7d','8d','9d','10d']
plt.xticks(x, labels[:len(x)])
plt.yticks(y, labels[:len(y)])
plt.tick_params(axis='x',direction='inout',length = 10, pad = 5, labelsize = 10, labelcolor = 'red', right=True)
plt.tick_params(axis='y',direction='out', length = 5, pad = 10, labelsize = 5, color = 'g', width = 3)
plt.title('Title', color='b',loc='left', pad=15, fontsize = 20, fontweight= 'bold' )
plt.show()
```

#수직 수평선 그리기
수평선 -axhline() ,hlines()
수직선 -axvline(), vlines()

```python
import numpy as np
import matplotlib.pyplot as plt


plt.plot(x,x, 'k*', linewidth = 15)
plt.plot(x,x**2, 'b^-.')
plt.plot(x,x**3, 'ro-', markersize = 15)

plt.axhline(17.5, 0.1, 0.9, linestyle='-.', color='yellow', linewidth=6)
plt.hlines(17.5, 0.1, 0.9, linestyle='-', color='royalblue', linewidth=9)
plt.axvline(0,0.2, 0.7, linestyle='-.', color='green', linewidth=2)

plt.vlines(0,10.2,10.7, linestyle='--', color = 'red', linewidth=2)
plt.show()
```
