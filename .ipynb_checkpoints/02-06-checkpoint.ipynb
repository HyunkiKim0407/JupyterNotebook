```python
plt.plot([1,2,3,4],[1,4,9,16])
#폰트 스타일: fontdict={'family':, 'color': ,'weight':, 'size':}
font1 ={'family':'serif', 'color':'b' ,'weight':'bold', 'size':15}
font2 ={'family':'serif', 'color':'deeppink' ,'weight':'normal', 'size':'xx-large'}
plt.xlabel('X_Label',labelpad = 50, fontdict = font1, loc = 'right')#x축  : left, right, center
plt.ylabel('Y_Label', labelpad = 20, fontdict = font2, loc = 'top')#y축 : top,center, bottom
plt.show()
```

```python
#선종류
plt.plot([1,2,3],[1,1,1],'-',color = 'C0', label = 'Solid') #solid
plt.plot([1,2,3],[2,2,2],'--',color = 'C1', label = 'dashed')#dashed
plt.plot([1,2,3],[3,3,3],'.',color = 'C2', label = 'dotted')# dotted
plt.plot([1,2,3],[4,4,4],'-.',color = 'C3', label = 'dash-dot')# dash-dot

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.axis([0.8,3.2,0.5,4.5])
plt.legend(loc='best', ncol = 4)

plt.show()
```

```python
#선종류
# plt.plot([1,2,3],[1,1,1],linestyle = 'solid', color = 'C0', label = 'Solid') #solid
# plt.plot([1,2,3],[2,2,2],linestyle = 'dashed', color = 'C1', label = 'dashed')#dashed
# plt.plot([1,2,3],[3,3,3],linestyle = 'dotted', color = 'C2', label = 'dotted')# dotted
# plt.plot([1,2,3],[4,4,4],linestyle = 'dashdot', color = 'C3', label = 'dash-dot')# dash-dot



plt.plot([1,2,3],[1,1,1],linestyle = (0,(1,1)), color = 'C0', label = 'dot') #solid
plt.plot([1,2,3],[1.5,1.5,1.5],linestyle = (0,(1,5)), color = 'C1', label = 'dot')#dashed
plt.plot([1,2,3],[3,3,3],linestyle = (0,(5,1)),color = 'C2', label = 'dot')# dotted
plt.plot([1,2,3],[2.5,2.5,2.5],linestyle = (0,(3,5,1,5)), color = 'C3', label = 'dot')# dashdot
plt.plot([1,2,3],[4,4,4],linestyle = (0,(3,5,1,1,1,5)), color = 'C4', label = 'dot')# dashdot




plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.axis([0.8,3.2,0.5,4.5])
plt.legend(loc='best', ncol = 4)

plt.show()
```

```python
#선끝모양
plt.plot([1,2,3],[1,1,1],linestyle = 'solid',linewidth = 10,solid_capstyle = 'butt' , color = 'C0', label = 'solid + butt') #solid
plt.plot([1,2,3],[2,2,2],linestyle = 'solid', linewidth = 10,solid_capstyle = 'round' , color = 'C1', label = 'solid + round')#dashed
plt.plot([1,2,3],[3,3,3],linestyle = 'dashed', linewidth = 10,dash_capstyle = 'butt' , color = 'C2', label = 'dashed + butt')# dotted
plt.plot([1,2,3],[4,4,4],linestyle = 'dashed', linewidth = 10,dash_capstyle = 'round' , color = 'C3', label = 'dashed + round')# dash-dot


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.axis([0.8,3.2,0.5,4.5])
plt.legend(loc='upper center', ncol = 4, frameon = False)

plt.show()
```
